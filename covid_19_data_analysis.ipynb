# droping the unnecessary columns

```python
covid_df=covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
covid_df.head()
```

## Active cases

```python
covid_df['Active_Cases']=covid_df['Confirmed']-(covid_df['Cured']+ covid_df['Deaths'])
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

# # Creating a custom table for better understanding of the data 

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory')
```

```python
statewise['Recovery Rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```


# Top 10 active and cured cases states

```python
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases','Date'],ascending=False).reset_index()
```

```python
top_10_active_cases
```

```python
top_10_Cured_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Cured','Date']].sort_values(by=['Cured','Date'],ascending=False).reset_index()
```

```python
top_10_Cured_cases
```

# remove the row without statesname

```python
index_names = covid_df[covid_df['State/UnionTerritory'] == 'Cases being reassigned to states' ].index
covid_df.drop(index_names,inplace=True)
```

```python
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases','Date'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 states with most active cases in India",size=25)
ax=sns.barplot(data=top_10_active_cases[:10],y="Active_Cases",x="State/UnionTerritory",linewidth=3,edgecolor='red')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

# Top states with highest deaths

```python
top_10_deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths','Date'],ascending=False).reset_index()
```

```python
top_10_deaths
```

```python
top_10_deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths','Date'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 states with most deaths in India",size=25)
ax=sns.barplot(data=top_10_deaths[:12],y="Deaths",x="State/UnionTerritory",linewidth=3,edgecolor='black')
plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```
